```python
import numpy as np

def my_sin(x):
    return np.sin(x)
```

```python
def my_cos(x):
    return np.cos(x)
```

```python
x = np.linspace(0, 2, 1000)
sin_vals = my_sin(x)
cos_vals = my_cos(x)
```

```python
print(f'{"x":>10} {"sin(x)":>12} {"cos(x)":>12}')

for i in range(10):
    print(f'{x[i]:10.6f} {sin_vals[i]:12.6f} {cos_vals[i]:12.6f}')
```
